### 1. Carregue o dataset e exiba as 10 primeiras linhas.

````python
df = pd.read_csv("./dataset/household_power_consumption.txt", sep=";")

df.head(10)
````

### 2. Explique a diferença entre as variáveis Global_active_power e Global_reactive_power.

A variável ```gloabal_active_power``` se refere a potência ativa média por minuto sendo utilizada da casa inteira (quilowatts), é a potência que está sendo convertida em trabalho útil.

A variável ```global_reactive_power``` se refere a potência reativa média por minuto sendo utiliziada da casa inteira (quilowatts), é a potência que mantém os campos magnéticos em equipamentos que precisam.

### 3. Verifique se existem valores ausentes no dataset. Quantifique-os.

````python
df.replace("?", np.nan, inplace=True) # Função que substitui os valores ausentes por NaN 

empty_after = df.isna() # Uma cópia do dataframe onde cada cada célula possui um true ou um false para NaN
print(empty_after.sum()) # Exibe a quantidade de NaN em cada coluna (contabilizando os trues)

df.head(20)
````

### 4. Converta a coluna Date para o tipo datetime e crie uma nova coluna com o dia da semana correspondente.

````python
df.info()

df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

df["Date"]
````

### 5. Filtre os registros apenas do ano de 2007 e calcule a média de consumo diário de Global_active_power.

````python
rows_year2007 = df[(df["Date"].dt.year) == 2007]

print(f"A média de consumo diário no ano de 2007 foi de: {round(rows_year2007['Global_active_power'].mean(), 2)} kW")
````

### 6. Gere um gráfico de linha mostrando a variação de Global_active_power em um único dia à sua escolha.

````python
df["Global_active_power"] = pd.to_numeric(df["Global_active_power"])

day_17_12_2006 = df[df["Date"] == "2006-12-17"]

day_17_12_2006.plot(x = "Time", y ="Global_active_power", kind='line', figsize= (16, 6))
plt.xlabel("Hora do Dia")
plt.ylabel("Potência Ativa (kW)")
plt.show()
````

### 7. Crie um histograma da variável Voltage. O que pode ser observado sobre sua distribuição?

````python
df['Voltage'] = pd.to_numeric(df['Voltage'])

plt.hist(df["Voltage"].dropna(), bins=50, edgecolor="black")
plt.xlabel("Voltage")
plt.ylabel("Frequência")
plt.title("Distribuição de Voltage")
plt.show()
````

A maior parte dos registros indicam que a voltagem ficou entre 235 e 245 volts. Isso indica que a casa teve uma boa estabilidade de tensão considerando o longo período de registros, e além disso não houveram tantas anomalias significativas que fizessem a voltagem passar de 250 volts ou menor que 230 volts.

### 8. Calcule o consumo médio por mês em todo o período disponível no dataset.

````python
df["YearMonth"] = df["Date"].dt.to_period("M")
monthly_mean = df.groupby("YearMonth")["Global_active_power"].mean()

mean_all_months = monthly_mean.mean()

print(f"O consumo médio por mês durante todo período foi de: {round(mean_all_months, 2)} kW ou {1000 * round(mean_all_months, 5)} W")
````

### 9. Identifique o dia com maior consumo de energia ativa global (Global_active_power).

````python
daily_sum = df.groupby(df["Date"].dt.date)["Global_active_power"].sum()
max_day = daily_sum.idxmax()
max_value = daily_sum.max()

print("Dia com maior consumo:", max_day, "->", max_value)
````

### 10. Compare o consumo médio de energia ativa global em dias de semana versus finais de semana.

````python
df["Weekday"] = df["Date"].dt.weekday # Segunda - 0, Terça - 1, ..., Domingo - 6

df["is_weekend"] = df["Weekday"] >= 5

weekday_mean = df[df["is_weekend"]]["Global_active_power"].mean()
weekend_mean = df[~df["is_weekend"]]["Global_active_power"].mean()

print("Semana:", weekday_mean, " | Fim de semana:", weekend_mean)
````
